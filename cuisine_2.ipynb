```python
# Read the multi-line string from file or directly
with open('unique_cuisines.txt', 'r') as f:
    lines = f.readlines()

# Use a set to collect unique individual cuisines
unique_countries = set()

for line in lines:
    # Split by comma, strip spaces
    parts = [c.strip() for c in line.strip().split(',')]
    unique_countries.update(parts)

# Sort for cleaner output
unique_countries_sorted = sorted(unique_countries)

# Print and save
print(f"Total unique cuisines/countries: {len(unique_countries_sorted)}\n")
for cuisine in unique_countries_sorted:
    print(cuisine)

# Save to file
with open('unique_individual_cuisines.txt', 'w') as f:
    for cuisine in unique_countries_sorted:
        f.write(cuisine + '\n')
```
